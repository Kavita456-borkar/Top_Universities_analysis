### // Country Performance:
### 🎯Which countries have the most universities in the top 100 rankings?

```python
plt.figure(figsize=(20,9))
most_universities = df[df['world_rank'] <= 100].drop_duplicates('institution').groupby('country')['institution'].count().sort_values(ascending=False)
plt.barh(most_universities.index, most_universities.values, color='mediumseagreen', alpha=0.8)
plt.title("Countries with Top 100 Ranking Universities", fontsize=16)
plt.xlabel("Number of Universities", fontsize=12)
plt.ylabel("Country", fontsize=12)
plt.gca().invert_yaxis() 

plt.tight_layout()
plt.show()
```

### // Factor Analysis:

#### 🎯What factors contribute most to a university's overall ranking?

```python
ranking_columns = ['world_rank', 'quality_of_education', 'alumni_employment', 'score','quality_of_faculty']
correlation = df[ranking_columns].corr()
correlation
```

```python
jd = df.corr(numeric_only=True)['score'].drop('score').sort_values(ascending=False)
jd

plt.figure(figsize=(20, 9))
sb.barplot(x=jd.values, y=jd.index, palette='viridis')
plt.title("Correlation of Factors with Overall Score", fontsize = 14)
plt.xlabel("Correlation Coefficient")
plt.ylabel("Ranking Factors")
plt.show()
```

#### The strongest correlations are with quality_of_faculty, quality_of_education, and world_rank, suggesting these have the most significant inverse relationships with the target variable .Weaker correlations (e.g., national_rank, year) indicate less influence on the target variable.

### 🎯 How does research impact (Citations, Publications, Broad Impact) influence rankings?

```python
factors = ['citations', 'publications', 'broad_impact']
colors = ['skyblue', 'salmon', 'mediumseagreen']

plt.figure(figsize=(20, 7))

for i, factor in enumerate(factors):
    plt.subplot(1, 3, i+1)
    sb.scatterplot(data=df, x=factor, y='world_rank', color=colors[i], s=70, alpha=0.8)
    plt.title(f'{factor} vs. World Rank', fontsize=14)
    plt.xlabel(factor)
    plt.ylabel('World Rank')

plt.tight_layout()
plt.show()

```

 #### The broad impact seems to have the strongest link to world rank in comparison with other two 

### 🎯 How do Quality of Education, Alumni Employment, and Faculty Quality correlate with rankings?

```python
factors = ['quality_of_education', 'alumni_employment', 'quality_of_faculty']
colors = ['skyblue', 'salmon', 'mediumseagreen']

plt.figure(figsize=(20, 9))

for i, factor in enumerate(factors):
    plt.subplot(1, 3, i+1)
    sb.scatterplot(data=df, x=factor, y='score', color=colors[i], s=70, alpha=0.8)
    plt.title(f'{factor} vs. Score', fontsize=14)
    plt.xlabel(factor)
    plt.ylabel('Score')

plt.tight_layout()
plt.show()
```

#### All three factors— Quality of Education, Alumni Employment, and Quality of Faculty— show a negative correlation with world_rank. This means better scores in these areas are typically associated with better university rankings. The patterns suggest that improvements in any of these metrics can contribute positively to a university's global standing.
#### However, the relationship is more scattered compared to metrics like citations or broad impact, meaning these factors may influence ranking but not as strongly.

### 🎯How have university rankings changed over time?

```python
df.groupby('year')['world_rank'].mean()
```

#### In 2012 and 2013, the average world rank was 50.5, indicating that the dataset primarily included top-ranked universities during those years In 2014 and 2015, the average world rank jumped to 500.5, suggesting a significant expansion in the number of universities included, covering a much broader range of rankings, including lower-ranked institutions.
#### The average world rank increased, not because universities are performing worse, but because more institutions (with lower ranks) were considered.

### 🎯Are there universities that consistently improve or decline in rankings?

```python
pivot = pd.pivot_table(data=df, index="year", columns="institution", values="world_rank", aggfunc='sum')
pivot
```

```python
yt = df.pivot(index= 'institution', columns= 'year', values= 'world_rank')

yt = yt.dropna(thresh=3) 

yt['Mean Rank'] = yt.mean(axis=1)
yt = yt.sort_values('Mean Rank').drop(columns ='Mean Rank')

plt.figure(figsize=(14, 10))
sb.heatmap(yt, annot=True, fmt=".0f", cmap="YlGnBu", linewidths=0.7, linecolor='gray', cbar_kws={'label': 'Rank'})
plt.title("University Ranking Trends Over Time (Lower = Better)",fontsize = 14)
plt.xlabel("Year")
plt.ylabel("University")
plt.gca().invert_yaxis() 
plt.tight_layout()
plt.show()
```

#### Yes, there are institutions which are consistantly showing growth in their rank. 
#### New York University ranking trend. In 2012-23 and in 2015-18, it shows consistant growth.
#### Let's see Osaka University ranking trend. In 2012-32 and in 2015-454, it shows decline in rank.
